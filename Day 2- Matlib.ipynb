```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x=[2,3,5,6]
y=[1,3,4,5]

plt.plot(x,y)
plt.show()
```

```python
x=[5,8,6,6]
y=[6,9,8,2]

plt.plot(x,y)
plt.title('line graph')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.show()
```

```python
x1=[5,8,9,12]
y1=[1,2,3,4]

x2=[1,2,3,4]
y2=[5,8,9,12]
plt.plot(x1,y1, label='line1') #label to assign the name to the line

plt.plot(x2,y2, label='line2')
plt.legend() #to show the labels on the graph
plt.grid() #to shpw square grid
plt.xlabel('x-axis') #to show x axis label
plt.ylabel('y-axis') #to show y axis label
plt.show() #to print the graph
```

```python
x=[2,3,5,6]
y=[1,3,4,5]

plt.plot(x,y,color='magenta',marker='s', markersize=10,linewidth=5, linestyle='--', label='First line')
plt.plot(y,x,'green',linewidth=5,linestyle=':',label='Second line')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
x=[2,3,5,6]
y=[1,3,4,5]

plt.plot(x,y,'b--*',markersize=15,linewidth=6)
plt.show()
```

```python
x=np.arange(0,10,0.1)
x1=np.sin(x)
y1=np.cos(x)
z1=np.tan(x)
#plt.plot(x1,'c--',label='Sin line',linewidth=5)
#plt.plot(y1,'m-',label='Cos line',linewidth=5)
plt.plot(z1,'b-',label='Cos line',linewidth=5)
plt.title('Sin Cos Graph',size=10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
x=[2,5,7,9]
y=[12,15,16,8]

plt.bar(x,y, label='Data')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
x1=[3,5,7,9]
y1=[12,15,16,8]
x2=[4,6,8,10]
y2=[15,17,19,10]
plt.bar(x1,y1,label='Data 1')
plt.bar(x2,y2,label='Data 2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
x1=[3,5,7,9]
y1=[12,15,16,8]
x2=[4,6,8,10]
y2=[15,17,19,10]
plt.barh(x1,y1,label='Data 1')
plt.barh(x2,y2,label='Data 2')
plt.title('Horizontal Bar')
plt.xlabel('Frequency')
plt.ylabel('Categorical')
plt.legend()
plt.show()
```

```python
Company=['Infosys','TCS','HCL','Wipro','Google','FB','Amazon','Twitter']
dframe=DataFrame({'Revenue(in$million)':[200,240,300,140,600,770,250,450]}, index=Company)
```

```python
dframe
```

```python
y=dframe.values
print(y)
y.flatten()
```

```python
y=dframe.values
y=y.flatten()
xpos=np.arange(8)
plt.bar(Company,y,label='Revenue')
#plt.xticks(xpos,Company)
plt.ylabel('Revenue in Million')
plt.ylabel('Companies')
plt.title('Companies Bar')
plt.legend()
plt.show()
```

```python
y=dframe.values
y=y.flatten()
xpos=np.arange(8)
plt.barh(xpos,y,label='Revenue')
plt.yticks(xpos,Company)
plt.ylabel('Revenue in Million')
plt.ylabel('Companies')
plt.title('Companies Bar')
plt.legend()
plt.show()
```

```python
expense_value=[1400,600,300,410,250]
expense_labels=['Home Rent','Food','Phone Bill','Car Maintainance','other utilities']

plt.pie(expense_value, labels=expense_labels, autopct='%1.1f%%', radius=1.5,explode=(0,0,0,0,0.8))
plt.title('Pie Chart')
plt.axis('equal')
plt.legend()
plt.savefig('piechart.jpg')
plt.show()
```
