```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
tips.info()
```

```python
tips.nunique()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, scatter=True, fit_reg=False, height=8)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, hue='size',row='day',scatter=True, fit_reg=False, height=8)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, hue='smoker',scatter=True,col='time', fit_reg=False, height=6, palette='autumn',markers=['o','x'])
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, hue='smoker',scatter=True,col='time',row='sex', fit_reg=False, height=6, palette='autumn',markers=['o','x'])
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='rainbow',size=6,jitter=False)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,palette='inferno',size=6)
plt.show()
```

```python
sns.countplot(x='size',data=tips,palette='inferno')
plt.show()
```

```python
sns.countplot(x='sex', hue='time',data=tips,palette='inferno')
plt.show()
```

```python
sns.countplot(x='day',data=tips,palette='inferno')
plt.show()
```

```python
sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,size=6,kind='hex')
plt.show()
```

```python
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,whis=False)
plt.show()
```

```python
sns.violinplot(x='total_bill',y='day',data=tips)
plt.show()
```

```python
sns.violinplot(x='total_bill',data=tips)
plt.show()
```

```python
sns.factorplot("size","total_bill",data=tips,hue="day");
```
