```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.DataFrame({'sales':[50,70,88,112,120,135,150], 'price':[7.5,6.0,5.8,5.0,4.5,4.0,3.7],'ad_cost':[35,40,45,50,55,68,75]})
```

```python
sns.pairplot(df, diag_kind='kde')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
model = LinearRegression()
```

```python
x= pd.DataFrame(df.price)
y= df.sales
```

```python
type(y)
```

```python
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
ypred=model.predict(x)
```

```python
model.score(x,y)
```

```python
np.sqrt(np.mean((ypred-y)**2))
```
